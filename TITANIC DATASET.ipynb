# Set a general style

```python
sns.set_style('whitegrid')
sns.set_style('darkgrid')
sns.set_style('ticks')
sns.set_style('white')
sns.set_style('dark')
```

# Pairplot to see relationships

```python
print("\nGenerating Pairplot...")
sns.pairplot(data, hue='Survived', diag_kind='kde')
plt.suptitle('Pairplot of Titanic Dataset', y=1.02)
plt.show()
```

# Heatmap of correlations

```python
# Select only numeric columns
numeric_df = data.select_dtypes(include=['number'])

# Now create heatmap safely
plt.figure(figsize=(10,7))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

# Histograms for numerical features

```python
print("\nPlotting Histograms...")
num_cols = ['Age', 'Fare', 'SibSp', 'Parch']
data[num_cols].hist(bins=15, figsize=(15, 6), layout=(2, 2), color='teal')
plt.suptitle('Histograms of Numerical Features')
plt.show()
```

# Boxplots to check outliers

```python
print("\nPlotting Boxplots...")
plt.figure(figsize=(15,8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=data[col], color='lightblue')
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

# Scatterplots

```python
print("\nPlotting Scatterplots...")
plt.figure(figsize=(15,8))
```

# Scatter Age vs Fare, colored by Survived

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=data, palette='viridis')
plt.title('Scatterplot of Age vs Fare by Survival')
plt.show()
```
